````python
import requests
import json
from pprint import pprint
````

````python
def request_info(url: str):
    r = requests.get(url)
    return r.json()
````

````python
def get_auction_data(page: int):
    first_page = request_info(f"https://api.hypixel.net/skyblock/auctions?page={page}")
    auction_data = first_page.get("auctions")
    return auction_data
````
